## Wav Conversion and Applying Filter(Noise Reduction)

```python
# Converting all the audio files into the .wav format
import os
import librosa
import soundfile as sf
import noisereduce as nr
from pydub import AudioSegment # will convert the format
```

```python
# defining the path for the input and output(converted wav will go here)
input_dir = "./genrenew"
output_dir = "cleaned_songs_wav"
genres = ['bollypop', 'carnatic', 'ghazal', 'semiclassical', 'sufi']
```

```python
# Foramt Conversion and Noise reduction
def format_converse_noise_reduce(input_dir, output_dir, genres):
    for genre in genres:
        input_genre_folder = os.path.join(input_dir,genre)
        out_genre_folder = os.path.join(output_dir,genre)
        os.makedirs(out_genre_folder, exist_ok=True) #making this genre folder in the cleaned_songs_wav

        print(f"--- Processing Genre: {genre} ---")

        for file_name in os.listdir(input_genre_folder):
            if file_name.endswith(('.mp3', '.m4a', '.wav')):

                try:
                    # 1. convert to wav using pydub
                    file_path = os.path.join(input_genre_folder,file_name)
                    # changing in a format which python can read and manipulate the audio by audiosegment
                    #mp3's is in compress format 
                    # by converting it calles ffmpeg(installed in the system) which decompress the audio for manipulation by python
                    # It loads the entire song into your system's memory (RAM) as an AudioSegment object.

                    #here audio is the audio segment object
                    audio = AudioSegment.from_file(file_path)

                    # making the temporary wav file for librosa
                    temp_wav = "temp.wav" #temp file name
                    audio.export(temp_wav, format = "wav")


                    # load with librosa for noise reduction
                    y,sr = librosa.load(temp_wav, sr = None)

                    # applying the noise reduction
                    # stationary=True is best for background hiss in old Hindi songs
                    reduced_noise = nr.reduce_noise(y= y, sr = sr, stationary=True) #stationary  = true for background hiss sound in old hindi songs

                    # save cleaned file
                    clean_file_name = os.path.splitext(file_name)[0] + ".wav"
                    final_output_path = os.path.join(out_genre_folder,clean_file_name) #conatains the file where reduced noise file will be written
                    sf.write(final_output_path, reduced_noise, sr)

                    print(f"Cleaned: {clean_file_name}")
                
                except Exception as e:
                    print(f"Error processing: {file_name}: {e}")  #file name is the original file name before conversion to wav

```

```python
format_converse_noise_reduce(input_dir, output_dir, genres)
```

## Format conversion only

```python
# defining the path for the input and output(converted wav will go here)
input_dir = "./genrenew"
output_dir = "cleaned_songs_only_wavs"
genres = ['bollypop', 'carnatic', 'ghazal', 'semiclassical', 'sufi']
```

```python
# Format Conversion Only
def format_converse(input_dir, output_dir, genres):
    for genre in genres:
        input_genre_folder = os.path.join(input_dir,genre)
        out_genre_folder = os.path.join(output_dir,genre)
        os.makedirs(out_genre_folder, exist_ok=True) #making this genre folder in the cleaned_songs_wav

        print(f"--- Converting Genre: {genre} ---")

        for file_name in os.listdir(input_genre_folder):
            if file_name.endswith(('.mp3', '.m4a', '.wav')):

                try:
                    # 1. convert to wav using pydub
                    file_path = os.path.join(input_genre_folder,file_name)
                    # changing in a format which python can read and manipulate the audio by audiosegment
                    #mp3's is in compress format 
                    # by converting it calles ffmpeg(installed in the system) which decompress the audio for manipulation by python
                    # It loads the entire song into your system's memory (RAM) as an AudioSegment object.

                    #here audio is the audio segment object

                    # 1. Load the compressed audio (mp3/m4a) into RAM
                    audio = AudioSegment.from_file(file_path) # the ffmpeg error is coming from here    

                    # 2. Define the new name (change extension to .wav)
                    clean_file_name = os.path.splitext(file_name)[0] + ".wav"
                    final_output_path = os.path.join(out_genre_folder,clean_file_name) #conatains the file where reduced noise file will be written
                    
                    # 3. Export directly to the destination folder 
                    audio.export(final_output_path, format = "wav")

                    print(f"Converted: {clean_file_name}")
                
                except Exception as e:
                    print(f"Error converting: {file_name}: {e}")  #file name is the original file name before conversion to wav

```

```python
import librosa
import soundfile as sf
import os

def format_converse_alt(input_dir, output_dir, genres):
    for genre in genres:
        input_genre_folder = os.path.join(input_dir, genre)
        out_genre_folder = os.path.join(output_dir, genre)
        os.makedirs(out_genre_folder, exist_ok=True)

        print(f"--- Converting Genre: {genre} ---")

        for file_name in os.listdir(input_genre_folder):
            if file_name.lower().endswith(('.mp3', '.m4a', '.wav')):
                try:
                    file_path = os.path.join(input_genre_folder, file_name)
                    
                    # 1. Use Librosa to load (Bypasses pydub's JSON parsing)
                    y, sr = librosa.load(file_path, sr=None)

                    # 2. Define output path
                    clean_file_name = os.path.splitext(file_name)[0] + ".wav"
                    final_output_path = os.path.join(out_genre_folder, clean_file_name)

                    # 3. Use soundfile to write (Very stable)
                    sf.write(final_output_path, y, sr)

                    print(f"Successfully Converted: {clean_file_name}")
                
                except Exception as e:
                    print(f"Error converting {file_name}: {e}")
```
