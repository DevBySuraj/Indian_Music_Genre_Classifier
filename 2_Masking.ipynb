# Masking(Source Separation into 4 stems By Spleeter)

```python
import os
from spleeter.separator import Separator
# initializing the Separator of 4 stems
separator = Separator('spleeter:4stems')
```

```python
# defining the path for the input and output(converted wav will go here)
input_dir = "./cleaned_songs_only_wavs" #- the wav ones without the noise reduce
output_dir = "hindi_stems"  
genres = ['bollypop', 'carnatic', 'ghazal', 'semiclassical', 'sufi']
```

```python
def source_separator(input_dir, output_dir, genres):
    for genre in genres:
        input_genre_folder = os.path.join(input_dir,genre)
        output_genre_folder = os.path.join(output_dir,genre)

        #creating the genre folder in the hindi_stems(output dir)
        if not os.path.exists(output_genre_folder):
            os.makedirs(output_genre_folder)

        print(f"--- Processing Genre: {genre} ---")

        for file_name in os.listdir(input_genre_folder):
            if file_name.endswith(".wav"):
                file_path = os.path.join(input_genre_folder,file_name)

                print(f"Splitting slots for: {file_name}")

                # 2. Separate the audio
                # This will create a subfolder named after the song containing:
                # vocals.wav, drums.wav, bass.wav, other.wav
                separator.separate_to_file(file_path, output_genre_folder)

    print("✅ All 500 songs have been separated into 4-stem slots!")
```

```python
# --- STEP 1: FORCE PATH AT THE TOP ---
# This ensures that even background processes see your clean FFmpeg
ffmpeg_bin = r'C:\FFmpeg\ffmpeg-8.0.1-full_build\bin'
os.environ["PATH"] = ffmpeg_bin + os.pathsep + os.environ["PATH"]
source_separator(input_dir, output_dir, genres)
```
